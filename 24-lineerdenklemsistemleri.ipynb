# 🔥 Lineer Denklem Sistemleri: Gerçekliği Parçalara Ayırmak

Lineer denklem sistemleri, birden fazla lineer denklemin bir araya gelmesiyle oluşan ve aynı değişkenleri içeren denklem kümeleridir. Bu sistemlerin çözümleri, tüm denklemleri aynı anda sağlayan değişken değerleridir. Lineer denklem sistemleri, matematik, fen bilimleri, mühendislik, ekonomi, bilgisayar bilimi ve daha birçok alanda yaygın olarak kullanılır.

**Genel Form:**

_m_ denklem ve _n_ bilinmeyenden oluşan genel bir lineer denklem sistemi şu şekilde ifade edilebilir:

```
a₁₁x₁ + a₁₂x₂ + ... + a₁ₙxₙ = b₁
a₂₁x₁ + a₂₂x₂ + ... + a₂ₙxₙ = b₂
...
aₘ₁x₁ + aₘ₂x₂ + ... + aₘₙxₙ = bₘ
```

Burada:

- `x₁, x₂, ..., xₙ`: Bilinmeyenler (değişkenler)
- `aᵢⱼ`: Katsayılar (i. denklemin j. bilinmeyeninin katsayısı)
- `bᵢ`: Sabit terimler (i. denklemin sağ tarafındaki değer)




**Matris Gösterimi:**

Lineer denklem sistemleri, matrisler ve vektörler kullanılarak daha özlü bir şekilde ifade edilebilir:

```
Ax = b
```

Burada:

- `A`: Katsayılar matrisi (m x n boyutlu)
- `x`: Bilinmeyenler vektörü (n x 1 boyutlu)
- `b`: Sabit terimler vektörü (m x 1 boyutlu)

## 📐 Ne Demek Bu? Basitçe Açıklarsak:

Bir lineer denklem sistemi, **birden fazla denklemi aynı anda çözmeyi amaçlayan**, **aynı değişkenleri paylaşan** matematiksel bir modeldir.

Örneğin:

```
x + y = 5  
2x - y = 1
```

Bu iki denklem, aynı x ve y değişkenlerine sahip.  
Ama her biri farklı bir kural tanımlıyor.  
Ve sen, **her iki kurala da uyan tek bir çözüm** arıyorsun.

## 💡 Gerçek Hayat Örneği: Market Sahibi Kar Hesabı

Sen bir market sahibisin.  
Günde iki ürün satıyorsun: ekmek ve yoğurt.  
Toplamda 5 adet ürün satıyorsun.  
Ama toplam gelirin 11 TL.  
Ekmek 2 TL, yoğurt 3 TL’ye satılıyor.

Kaç ekmek, kaç yoğurt sattın?

Denklemler şöyle:

```
x + y = 5   → Toplam satış  
2x + 3y = 11 → Toplam gelir
```

### Çözüm:
- İlk denklemde `y = 5 - x` deriz.
- İkinci denklemde yerine koyarız:
```
2x + 3(5 - x) = 11  
2x + 15 - 3x = 11  
-x + 15 = 11  
x = 4  
→ y = 1
```

✅ Sonuç: 4 ekmek, 1 yoğurt sattın.

> “Matematik, seninle konuşur. Ancak senin de dinlemeye zihnini açman gerekir.”

## 🛠️ Üç Temel Çözüm Durumu Vardır

| Durum | Açıklama | Görsel Yorum |
|--------|----------|---------------|
| ✅ **Tek çözüm** | Tüm denklemler aynı noktada buluşur | ↗️ ↔️ ↘️ → tek kesişim noktası |
| ✅ **Sonsuz çözüm** | Denklemler çakışık doğrular üretir | ——————– ≡ ——————– |
| ❌ **Çözüm Yok** | Denklemler paraleldir, kesişmez | ⇄ ⇄ → asla karşılaşmayan doğrular |


Bir lineer denklem sisteminin üç olası çözüm durumu vardır:

1. **Tek Çözüm:** Sistem tutarlıdır ve tek bir çözüm kümesi vardır. Geometrik olarak, doğruların (veya düzlemlerin) tek bir noktada kesişmesi anlamına gelir.
2. **Sonsuz Çözüm:** Sistem tutarlıdır ve sonsuz sayıda çözüm kümesi vardır. Geometrik olarak, doğruların (veya düzlemlerin) çakışık (aynı) olması veya bir doğru boyunca kesişmesi anlamına gelir.
3. **Çözüm Yok:** Sistem tutarsızdır ve hiçbir çözüm kümesi yoktur. Geometrik olarak, doğruların (veya düzlemlerin) paralel olması ve kesişmemesi anlamına gelir.

## 🧮 Matris Gösterimi: Lineer Sistemlerin Dili

Lineer denklem sistemlerini anlamak için en güçlü araçlardan biri **matris gösterimidir.**

Klasik sistem:

```
x + y = 5  
2x - y = 1
```

Bu şu şekilde matris haline gelir:

$$
\begin{bmatrix}
1 & 1 \\
2 & -1
\end{bmatrix}
\cdot
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
5 \\
1
\end{bmatrix}
$$

Yani:
$$
Ax = b
$$

Burada:
- `A`: Katsayılar matrisi
- `x`: Bilinmeyenler vektörü
- `b`: Sabit terimler vektörü

## 🎯 Lineer Sistem Nasıl Çözülür?

Üç ana yöntem vardır.  
Hepsi farklı ama aynı sonuca götürür.

### 1. **Yerine Koyma Yöntemi**
Bir denklemi çöz, diğerinde yerine koy.

```
x + y = 5 → y = 5 - x  
2x - y = 1 → 2x - (5 - x) = 1 → 3x = 6 → x = 2
```

✅ Sonuç: `x = 2`, `y = 3`

### 2. **Yok Etme Yöntemi**

Bir değişkeni yok et!

```
x + y = 5  
2x - y = 1  
--------------  
3x     = 6 → x = 2  
y = 5 - x → y = 3
```

✅ Aynı sonucu bulduk.

### 3. **Matris Yöntemi ile Python’da Çözmek**

```python
import numpy as np

A = np.array([[1, 1], [2, -1]])
b = np.array([5, 1])

x = np.linalg.solve(A, b)
print(f"x = {x[0]}, y = {x[1]}")
```

✅ Python çıktı:

```
x = 2.0, y = 3.0
```


## ⚖️ Homojen mi, Değil mi?

Bunu bilmek, çözümün türü açısından büyük fark yaratır.

### ✅ Homojen Sistem:
Tüm sabit terimler **sıfırsa**, sistem homojendir.

```
x + y = 0  
2x - y = 0
```

👉 Bu sistemde çözüm: `(0, 0)` → aşikar çözüm

### ❌ Homojen Olmayan Sistem:
Sabit terimler sıfır değilse, sistem homojen değildir.

```
x + y = 5  
2x - y = 1
```

👉 Bu sistemde çözüm olabilir. Ya da olmayabilir.  
Ama kesinlik yoktur.  
Yani gerçeklik daha karmaşıktır.

## 🧪 Lineer Sistemleri Çözen Yaklaşımlar

| Yöntem | Açıklama |
|--------|-----------|
| **Yerine Koyma (Substitution)** | En basit, en eski yöntem |
| **Yok Etme (Elimination)** | Denklem sayısını azaltarak ilerler |
| **Gauss Eliminasyonu** | Satır işlemleriyle çözüm bulunur |
| **Cramer Kuralı** | Determinantla çözülür, küçük sistemler için ideal |
| **LU Ayrıştırması** | Büyük sistemlerde hız sağlar |
| **Matris Tersi** | A matrisi tersinirse, `x = A⁻¹b` ile çözülür |

📌 Her yöntemin avantajı ve dezavantajı var.  
En önemlisi: **hangi yöntemin senin soruna uygun olduğunu bilmek.**


**Lineer Denklem Sistemlerini Çözme Yöntemleri:**

1. **Yerine Koyma Yöntemi (Substitution):** Bir denklemdeki bir değişkeni çözüp diğer denklemlerde yerine koyarak bilinmeyen sayısını azaltma yöntemidir.
2. **Yok Etme Yöntemi (Elimination):** Denklemleri uygun katsayılarla çarpıp toplayarak veya çıkararak bir veya daha fazla değişkeni yok etme yöntemidir. Gauss eliminasyonu ve Gauss-Jordan eliminasyonu bu yöntemin sistematik uygulamalarıdır.


**Matris Yöntemleri:**

- **Matris Tersini Bulma:** Eğer katsayılar matrisi `A` kare matris ise ve tersi varsa (`A⁻¹`), çözüm `x = A⁻¹b` olarak bulunur.
- **Gauss Eliminasyonu:** Sistemin genişletilmiş matrisini (katsayılar matrisi ve sabit terimler vektörünün birleştirilmesiyle oluşan matris) satır işlemleri kullanarak üst üçgensel forma dönüştürme yöntemidir.
- **Gauss-Jordan Eliminasyonu:** Gauss eliminasyonuna benzerdir, ancak matris tamamen indirgenmiş satır eşelon formuna dönüştürülür. Bu, doğrudan çözümü verir.
- **LU Ayrıştırması:** Matris `A`'yı alt üçgensel bir matris `L` ve üst üçgensel bir matris `U`'nun çarpımı olarak ayrıştırma yöntemidir. Bu, sistemin daha verimli bir şekilde çözülmesini sağlar.
- **Cramer Kuralı:** Determinantlar kullanarak lineer denklem sistemlerini çözme yöntemidir. Küçük sistemler için kullanışlıdır, ancak büyük sistemler için hesaplama açısından verimsiz olabilir.
