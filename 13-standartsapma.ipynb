# 📏 Standart Sapma: Verinin Ortalamaya Olan Uzaklığını Gösteren Ölçü

## Giriş: Veri Ne Kadar Yayılmış?

Diyelim ki iki farklı sınıfta ortalama not aynı: **75 puan**.  
Ama birinde herkes 70-80 arasında alırken, diğerinde bazıları 40, bazıları 100 aldı.

İkisinin ortalaması aynı ama veri nasıl yayılmış? İşte tam burada devreye **standart sapma** giriyor.

> “Standart sapma, verinin ortalamadan ne kadar uzaklaştığını gösterir. Aynı ortalama bile olsa, dağılım farklıysa standart sapma bize bunu söyler.”  


Ortalamadan sapmayı gösterir

### 🔍   ** Standart Sapma Nedir?**

Standart sapma (**σ**), bir veri kümesindeki sayıların **ortalamadan ne kadar saptığını** gösteren matematiksel bir ölçüdür.

- **Düşük standart sapma** → Veriler birbirine yakın, homojen.
    
- **Yüksek standart sapma** → Veriler birbirinden uzak, heterojen.
    

Örneğin:

- **A Takımı:** 70, 72, 71, 73 (standart sapma küçük, skorlar birbirine yakın)
    
- **B Takımı:** 50, 60, 80, 90 (standart sapma büyük, skorlar dağınık)

## 🧮 Standart Sapma Nasıl Hesaplanır?

### 1. **Ortalamayı Bul**
Tüm sayıların toplamını veri sayısına böl.

### 2. **Her Sayının Ortalamadan Farkını Al**
Sayı - ortalama = sapma

### 3. **Sapmaların Karelerini Al**
Negatifleri pozitif yapmak için karesini alıyoruz.

### 4. **Varyansı Hesapla**
Karelerin ortalamasını al.  
Eğer **örneklemse**: `n - 1` ile böl  
Eğer **ana kütle**yse: `n` ile böl

### 5. **Karekökü Al**
Bu da bize **standart sapmayı** verir.



### **Standart Sapma Nasıl Hesaplanır?**

Adım adım gidelim:

#### **Adım 1: Aritmetik Ortalama Bulunur**

Önce verilerin ortalaması hesaplanır.

**Örnek Veri:** 5, 7, 3, 7

- **Toplam:** 5 + 7 + 3 + 7 = **22**
    
- **Ortalama (μ):** 22 / 4 = **5.5**
    

#### **Adım 2: Her Verinin Ortalamadan Sapması Bulunur**

Her bir veri noktasından ortalamayı çıkarıyoruz.

|Veri (xᵢ)|Sapma (xᵢ - μ)|
|---|---|
|5|5 - 5.5 = **-0.5**|
|7|7 - 5.5 = **1.5**|
|3|3 - 5.5 = **-2.5**|
|7|7 - 5.5 = **1.5**|

#### **Adım 3: Sapmaların Kareleri Alınır**

Negatif değerleri yok etmek için karesini alıyoruz.

|Sapma (xᵢ - μ)|Karesi (xᵢ - μ)²|
|---|---|
|-0.5|0.25|
|1.5|2.25|
|-2.5|6.25|
|1.5|2.25|

#### **Adım 4: Varyans Hesaplanır**

Karelerin toplamını veri sayısına (n) veya (n-1)'e böleriz.

- **Popülasyon varyansı (tüm veri için):**
    
    σ2=0.25+2.25+6.25+2.254=114=2.75σ2=40.25+2.25+6.25+2.25​=411​=2.75
- **Örneklem varyansı (tahmin için):**
    
    s2=114−1≈3.666...s2=4−111​≈3.666...

#### **Adım 5: Standart Sapma Bulunur (Karekök Alınır)**

- **Popülasyon standart sapması:**
    
    σ=2.75≈1.66σ=2.75​≈1.66
- **Örneklem standart sapması:**
    
    s=3.666...≈1.91s=3.666...​≈1.91

Neden (n-1) Kullanılır?
Eğer veri tüm popülasyon ise n’e böleriz.
Eğer veri örneklem ise n-1’e böleriz (Bessel düzeltmesi).

Sebep: Örneklemde serbestlik derecesi azalır, daha doğru tahmin yapmak için (n-1) kullanılır.

## 💡 Gerçek Hayat Örneğiyle Anlamak

### 🎯 Sınıf Notları:

Notlar: `70, 80, 90, 60, 75`

Adım adım çözüm:

1. **Ortalama:**  
   $$
   \frac{70 + 80 + 90 + 60 + 75}{5} = 75
   $$

2. **Sapmalar:**  
   - 70 - 75 = -5  
   - 80 - 75 = +5  
   - 90 - 75 = +15  
   - 60 - 75 = -15  
   - 75 - 75 = 0

3. **Kareler Toplamı:**  
   - (-5)² = 25  
   - (+5)² = 25  
   - (+15)² = 225  
   - (-15)² = 225  
   - 0² = 0  
   → Toplam = **500**

4. **Varyans (Örneklem için):**  
   $$
   \frac{500}{5 - 1} = \frac{500}{4} = 125
   $$

5. **Standart Sapma:**  
   $$
   \sqrt{125} \approx 11.18
   $$

✅ Bu sınıfın standart sapması yaklaşık **11.18’dir.**  
Yani öğrencilerin notları ortalama olan 75’ten yaklaşık ±11 civarında değişiyor.


## 📊 Standart Sapmanın Özellikleri

| Özellik | Açıklama |
|--------|----------|
| ✅ Dağılım ölçüsüdür | Verinin yayılımını gösterir |
| ⚠️ Aykırı değerlere duyarlıdır | Çok yüksek veya düşük değerler sapmayı büyütür |
| ✅ Birim korur | Sayılarla aynı birime sahiptir (örneğin: puan, kg, TL) |
| 🟢 Varyansın kareköküdür | Varyans daha büyük sayılar verir, standart sapma daha kolay anlaşılır |


## 📈 Standart Sapma mı, Varyans mı?

| Özellik | Varyans | Standart Sapma |
|---------|---------|----------------|
| Formülü | Sapmaların karesi ortalaması | Varyansın karekökü |
| Birimi | Kareli (örneğin: puan²) | Aynı birim (örneğin: puan) |
| Yorumlanabilirlik | Zor | Kolay |
| Kullanım Alanı | İstatistik modellerde | Günlük analizlerde |

📌 **Özet:**  
- Varyans teknik olarak önemlidir ama  
- **Standart sapma günlük dilde daha çok kullanılır**, çünkü anlamlıdır.




## 📌 Nerelerde Kullanılır?

| Alan | Kullanım Yeri |
|------|----------------|
| **Finans** | Hisse senedi fiyatlarının dalgalanması |
| **Eğitim** | Öğrenci notlarının dağılımı |
| **Kalite Kontrolü** | Ürün boyutlarındaki sapmalar |
| **Sağlık** | Tıbbi verilerin dağılımı |
| **Veri Bilimi** | Özelliklerin dağılımı ve standardizasyonu |


```python
# Python Uygulaması: Standart Sapma Nasıl Hesaplanır?

import numpy as np

veri = [70, 80, 90, 60, 75]

# Örneklem standart sapması (n-1'e göre)
std_orneklem = np.std(veri, ddof=1)
print(f"Örneklem Standart Sapma: {std_orneklem:.2f}")

# Ana kütle standart sapması (n'e göre)
std_kitle = np.std(veri)
print(f"Ana Kütle Standart Sapma: {std_kitle:.2f}")
```

📌 `ddof=1`: örneklem için farkı hesaplamayı sağlar  
📌 `np.var()`: varyansı hesaplamak için kullanılabilir

Bir sınıftaki 5 öğrencinin boyları (cm): 160, 165, 170, 155, 175. Standart sapmasını hesaplayınız.

Ortalama = (160+165+170+155+175)/5 = 165

Varyans = (25+0+25+100+100)/5 = 50
