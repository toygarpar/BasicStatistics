# 📊 Verileri Grafiklerle Özetlemek: Görsel Anlatımın Gücü

## Giriş: Sayılar Sessiz Kalabilir, Görseller Konuşur!

Diyelim ki 100 satırlık bir veri setin var. Onu okumak zorunda mısın? Yoksa onu görselleştirip hemen anlam çıkarabilmeli misin?

İşte tam burada devreye **veri görselleştirme** girer.  
Sayıları değil, **hikâyeyi anlatmak** için kullanılır.

> “Veri dünyasında görseller, sayılardan daha güçlü konuşur. Çünkü herkes sayıları anlamlandıramaz ama herkes görsel anlayabilir.”  

## 🔍 Neden Grafik Kullanırız?

| Sebep | Açıklama |
|--------|----------|
| ✅ Anlaşılabilirlik | Sayılar karmaşık olabilir, grafikler basit ve net |
| ✅ Karşılaştırma Kolaylığı | Hangisi daha fazla, hangisi artıyor, hangisi azalıyor? Hemen görebilirsin |
| ✅ Eğilimleri Görmek | Zaman içindeki değişimleri izlemek için çok güçlüdür |
| ✅ Dikkat Çekmek | Sunumda, raporda ya da analizde ilk dikkati çeken şey grafiktir |

📌 Veriyle çalışırken en güçlü silah, **görsel temsil yeteneğidir.**

## 📈 Alan Grafiği: Değişimi ve Katkıyı Aynı Anda Görmek

### Nedir?
Alan grafiği, **çizgi grafik** ile **alan dolgusu** arasında bir gösterimdir. Özellikle zaman serisi verilerinde kullanılır.

### Ne İşe Yarar?
- Zaman içindeki eğilimleri gösterir
- Farklı kategorilerin toplam içindeki katkısını vurgular

### Türleri:
| Tür | Açıklama |
|-----|-----------|
| **Düz Alan Grafiği** | Her kategori bağımsız gösterilir |
| **Yığılmış Alan Grafiği** | Toplam katkılar üst üste sıralanır |
| **%100 Yığılmış Alan Grafiği** | Her zaman noktası %100’e göre orantılanır → oranlar ön planda |

### 💡 Gerçek Hayat Örneği:

Bir şirketin aylık satışları:

```
A Ürünü: [100, 120, 130, 150]  
B Ürünü: [80, 90, 110, 140]  
C Ürünü: [60, 70, 80, 100]
```

👉 Bu veriyi yığılmış alan grafiğine dökersek:
- Hangi ürünün katkı arttı?
- Toplam satış trendi nasıl gidiyor?
- Hangi aydaki artış daha belirgin?

Bu sorulara tek bakışta yanıt bulabilirsin.

## 📊 Çubuk Grafik: Kategoriler Arası Farkı Gösteren Silah

### Nedir?
Çubuk grafik, farklı **kategorileri karşılaştırmak** için kullanılan güçlü bir görsel araçtır.

### Ne İşe Yarar?
- Farklı şehirlerdeki satış rakamları
- Aylara göre gelir-gider karşılaştırması
- Ürün türlerinin performansları

### Türleri:
| Tür | Açıklama |
|------|----------|
| **Dikey Çubuk (Sütun)** | Genellikle aylık veriler için |
| **Yatay Çubuk** | Uzun kategori isimleri varsa tercih edilir |
| **Yığılmış Çubuk** | Her çubuk birden fazla bileşenden oluşur |
| **Gruplandırılmış Çubuk** | Aynı kategori için birden fazla veri seti varsa kullanılır |

### 💡 Gerçek Hayat Örneği:

Ürün satışları:

| Ay | A ürünü | B ürünü | C ürünü |
|----|---------|---------|---------|
| Ocak | 100     | 80      | 60      |
| Şubat | 120    | 90      | 70      |

👉 Bu verileri gruplanmış çubuk grafiğe dökersen:
“Hangi ürün hangi ayda öne çıktı?”  
sorusuna tek bakışta yanıt verebilirsin.

## 🥧 Pasta Grafiği: Bir Bütünün Parçalarını Göstermek

### Nedir?
Pasta grafiği, bir bütünü oluşturan parçaların **oranlarını** gösterebilen dairesel bir grafiktir.

### Ne İşe Yarar?
- Pazar payları
- Gelir kaynakları
- Anket sonuçları

📌 Ancak unutma:  
Pasta grafiği, **fazla kategori** olduğunda karışabilir. 5’ten fazla dilim yapmaktan kaçın.

### 💡 Gerçek Hayat Örneği:

Bir şirketin gelir kaynakları:

- Ürün Satışı: %50  
- Reklam: %30  
- Sponsorluk: %15  
- Diğer: %5  

👉 Bu veriyi pasta grafiğine döktüğünde:
“Kurum nereden para kazanıyor?”  
sorusuna tek bakışta yanıt alırsın.

## 📉 Çizgi Grafik: Trendleri Yakalamak İçin En Güçlü Silah

### Nedir?
Zaman içindeki değişimi göstermek için kullanılan çizgiyle oluşturulan grafiktir.

### Ne İşe Yarar?
- Hisse senedi fiyatlarında dalgalanmaları görmek
- Haftalık satışlarda yükselen/düşen eğilimleri yakalamak
- Sıcaklık değişimini zamana göre incelemek

### Türleri:
| Tür | Açıklama |
|------|----------|
| **Basit Çizgi Grafik** | Tek bir veri serisini gösterir |
| **Çoklu Çizgi Grafik** | Birden fazla seriyi aynı grafikte karşılaştırır |
| **Yığılmış Çizgi Grafik** | Seriler birbirinin üzerine eklenerek toplamı gösterir |

### 💡 Gerçek Hayat Örneği:

Aylık sıcaklık verisi:

```
Ocak: 5°C  
Şubat: 7°C  
Mart: 10°C  
Nisan: 15°C
```

👉 Bu verileri çizgi grafiğe döktüğünde:
“Sıcaklık artıyor mu, yoksa rastgele mi değişiyor?”  
sorusuna kolayca yanıt bulursun.


````python
# Python Uygulaması: Grafikler Nasıl Çizilir?

import matplotlib.pyplot as plt
import numpy as np

# Basit çizgi grafik
x = ["Ocak", "Şubat", "Mart", "Nisan"]
y = [5, 7, 10, 15]

plt.plot(x, y, marker='o')
plt.title("Sıcaklık Artışı")
plt.xlabel("Ay")
plt.ylabel("Sıcaklık (°C)")
plt.grid(True)
plt.show()
````

````python
# Yığılmış Alan Grafiği

import pandas as pd

data = {
    'A': [100, 120, 130, 150],
    'B': [80, 90, 110, 140],
    'C': [60, 70, 80, 100]
}
df = pd.DataFrame(data, index=["Ocak", "Şubat", "Mart", "Nisan"])

df.plot(kind='area', stacked=True)
plt.title("Yığmalı Alan Grafiği")
plt.xlabel("Ay")
plt.ylabel("Satışlar")
plt.show()
````

````python
# Gruplandırılmış Çubuk Grafik

import numpy as np

N = 4
ind = np.arange(N)  # x ekseni pozisyonları
width = 0.2        # çubuk genişliği

fig, ax = plt.subplots()

rects1 = ax.bar(ind, df['A'], width, label='A Ürünü')
rects2 = ax.bar(ind + width, df['B'], width, label='B Ürünü')
rects3 = ax.bar(ind + 2*width, df['C'], width, label='C Ürünü')

ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Ocak', 'Şubat', 'Mart', 'Nisan'))
ax.legend()
plt.title('Aylara Göre Ürün Satışları')
plt.show()
````

````python
# Pasta Grafiği

gelir_kaynaklari = [50, 30, 15, 5]
etiketler = ['Ürün Satışı', 'Reklam', 'Sponsorluk', 'Diğer']

plt.pie(gelir_kaynaklari, labels=etiketler, autopct='%1.1f%%')
plt.title("Gelir Kaynakları")
plt.axis('equal')  # Pasta grafiğini düzgün göster
plt.show()
````
