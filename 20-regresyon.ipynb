# Regresyon - Regression 


Regresyon tanımsal olarak bir bağımlı değişken ile bağımsız değişkenler arasındaki ilişkisini belirler.

Regresyon analizi bağımsız değişkenler değiştiğinde bağımlı değişkenin nasıl bir etki alacağını hesaplar.

Değişkenler arasında neden sonuç ilişkisini test eder.

---

Regresyon, istatistikte iki veya daha fazla değişken arasındaki ilişkiyi modellemek için kullanılan bir tekniktir. Özellikle bir bağımlı değişkenin (sonuç değişkeni olarak da bilinir- response, outcome) bir veya daha fazla bağımsız değişken (açıklayıcı değişkenler veya tahmin ediciler olarak da bilinir - predictor) tarafından nasıl etkilendiğini anlamak ve tahmin etmek için kullanılır.

# 📊 Regresyon: 

## Giriş: Birini Biliyorsan Diğerine Tahmin Yap!

Diyelim ki bir öğrencinin notlarını tahmin etmek istiyorsun. Onun ders çalışma süresini biliyorsun ama sınavda alacağı notu bilmiyorsun.

Ya da bir emlakçısın ve evin metrekaresine göre fiyatını kestirmek istiyorsun.

İşte tam burada devreye **regresyon** giriyor.  
Regresyon, bir değişkeni bilirsen, diğerini **nasıl tahmin edebileceğini gösteren yol haritasıdır.**

> “Regresyon, verideki neden-sonuç ilişkisini aramaktır. Bir şeyi bilince diğerine adım adım yaklaşılır.”  



## 🔍 Regresyon Nedir?

**Regresyon**, bir **bağımlı değişkeni** (Y) başka **bir veya birkaç bağımsız değişkenle** (X) açıklamaya veya tahmin etmeye çalışan matematiksel modeldir.

### Ana unsurlar:
| Terim | Açıklama |
|--------|----------|
| **Bağımlı Değişken (Y)** | Tahmin edilmeye çalışılan değişken → Örnek: Sınav notu, satış fiyatı, gelir |
| **Bağımsız Değişken(ler) (X)** | Kullanılan açıklayıcı değişken(ler) → Örnek: Çalışma süresi, ev boyutu, reklam bütçesi |
| **Katsayılar (β₀, β₁...):** | Modeldeki sabit terim ve eğimleri ifade eder |
| **Hata Terimi (ε):** | Modelin açıklayamadığı kısım |

📌 Basit doğrusal regresyon denklemi:
$$
Y = \beta_0 + \beta_1 X + \varepsilon
$$


## 💡 Gerçek Hayat Örneği: Ev Fiyatları

Bir emlak şirketi, ev fiyatlarını şu değişkenlerle tahmin etmek istiyor:

- Metrekare (X₁)
- Oda sayısı (X₂)
- Konum skoru (X₃)

👉 Bu durumda kullanacağımız model:
$$
\text{Ev Fiyatı} = \beta_0 + \beta_1 \cdot \text{Metrekare} + \beta_2 \cdot \text{Oda Sayısı} + \beta_3 \cdot \text{Konum Skoru} + \varepsilon
$$

📌 Eğer metrekare artarsa, fiyat da artıyor mu? Ya konum ne kadar etkili?

## 🎯 Neden Regresyon Kullanılır?

| Amaç | Açıklama |
|------|----------|
| **Tahmin Yapmak:** | Öğrencinin kaç alacağını, mağazada kaç ürün satılacağını, hisse senedinin değerini tahmin etmek için |
| **İlişkiyi Anlamak:** | Hangi faktörün daha çok etkili olduğunu görmek için |
| **Karar Almak:** | Veriye dayalı stratejiler oluşturmak için |

📌 Yani regresyon, **veriyle geleceği tahmin etmenin** en güçlü yollarından biridir.

## 🧮 Regresyon Türleri

## 📈 Basit Doğrusal Regresyon Nasıl Çalışır?

Varsayalım:
- Ders çalışma süresi (X): [2, 3, 4, 5, 6] saat/gün
- Sınav notu (Y): [50, 60, 70, 80, 90]

👉 Hedef: **“Kaç saat çalışırsa kaç puan alır?”** sorusuna yanıt vermek.

Model çıktısı şöyle olabilir:
$$
\text{Not} = 30 + 10 \cdot (\text{Çalışma Süresi})
$$

📌 Bu formüle göre:
- 2 saat çalışırsan: 30 + 10×2 = **50**
- 5 saat çalışırsan: 30 + 10×5 = **80**

Bu temel mantıkla **hiç görmediğin bir kişi için bile tahmin yapabilirsin.**

```python
# Python Uygulaması: Basit Doğrusal Regresyon

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Veri seti
x = np.array([2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([50, 60, 70, 80, 90])

# Regresyon modelini oluşturalım
model = LinearRegression()
model.fit(x, y)

# Eğim ve kesim noktası
print(f"Kesme (β₀): {model.intercept_:.2f}")
print(f"Eğim (β₁): {model.coef_[0]:.2f}")

# Tahmin yapalım
tahmin = model.predict(x)

# Grafiğe çizelim
plt.scatter(x, y, color='blue', label='Gerçek')
plt.plot(x, tahmin, color='red', label='Regresyon')
plt.title("Regresyon ile Tahmin")
plt.xlabel("Çalışma Saati")
plt.ylabel("Sınav Notu")
plt.legend()
plt.grid(True)
plt.show()
```

## 📉 Korelasyon mu, Regresyon mu?

| Özellik | Korelasyon | Regresyon |
|---------|-------------|------------|
| Ne ölçer? | İki değişken arasındaki ilişki gücü | Bağımlı değişkeni tahmin etme |
| Yönlü müdür? | Evet, pozitif/negatif olabilir | Evet, hangi yönde etki olduğunu gösterir |
| Tahmin yapar mı? | Hayır → sadece ilişki | Evet → yeni veri girdiğinde çıktı üretir |
| Nedensellik mi? | Hayır → ilişki kurar | Yok → ilişki belirtir ama kanıtlamaz |

📌 Korelasyon iki şeyin dans ettiğini/ilişkisini gösterir,  
✅ Regresyon ise onların dansının nasıl olduğunu/ilişkinin nedenselliğini söyler.

## 📐 R² (Belirlilik Katsayısı) Nedir?

R², modelin bağımlı değişkenin yüzde kaçını açıkladığını gösterir.

| Aralık | Yorum |
|---------|--------|
| **%100 → 1.00:** | Model mükemmel açıklıyor |
| **%70 → 0.70:** | Güçlü açıklama |
| **%50 → 0.50:** | Orta düzey açıklama |
| **%30 → 0.30:** | Zayıf açıklama |

📌 R², modelin gücünü gösteren en önemli göstergelerden biridir.
