# Regresyon - Regression 


Regresyon tanƒ±msal olarak bir baƒüƒ±mlƒ± deƒüi≈üken ile baƒüƒ±msƒ±z deƒüi≈ükenler arasƒ±ndaki ili≈ükisini belirler.

Regresyon analizi baƒüƒ±msƒ±z deƒüi≈ükenler deƒüi≈ütiƒüinde baƒüƒ±mlƒ± deƒüi≈ükenin nasƒ±l bir etki alacaƒüƒ±nƒ± hesaplar.

Deƒüi≈ükenler arasƒ±nda neden sonu√ß ili≈ükisini test eder.

---

Regresyon, istatistikte iki veya daha fazla deƒüi≈üken arasƒ±ndaki ili≈ükiyi modellemek i√ßin kullanƒ±lan bir tekniktir. √ñzellikle bir baƒüƒ±mlƒ± deƒüi≈ükenin (sonu√ß deƒüi≈ükeni olarak da bilinir- response, outcome) bir veya daha fazla baƒüƒ±msƒ±z deƒüi≈üken (a√ßƒ±klayƒ±cƒ± deƒüi≈ükenler veya tahmin ediciler olarak da bilinir - predictor) tarafƒ±ndan nasƒ±l etkilendiƒüini anlamak ve tahmin etmek i√ßin kullanƒ±lƒ±r.

# üìä Regresyon: 

## Giri≈ü: Birini Biliyorsan Diƒüerine Tahmin Yap!

Diyelim ki bir √∂ƒürencinin notlarƒ±nƒ± tahmin etmek istiyorsun. Onun ders √ßalƒ±≈üma s√ºresini biliyorsun ama sƒ±navda alacaƒüƒ± notu bilmiyorsun.

Ya da bir emlak√ßƒ±sƒ±n ve evin metrekaresine g√∂re fiyatƒ±nƒ± kestirmek istiyorsun.

ƒ∞≈üte tam burada devreye **regresyon** giriyor.  
Regresyon, bir deƒüi≈ükeni bilirsen, diƒüerini **nasƒ±l tahmin edebileceƒüini g√∂steren yol haritasƒ±dƒ±r.**

> ‚ÄúRegresyon, verideki neden-sonu√ß ili≈ükisini aramaktƒ±r. Bir ≈üeyi bilince diƒüerine adƒ±m adƒ±m yakla≈üƒ±lƒ±r.‚Äù  



## üîç Regresyon Nedir?

**Regresyon**, bir **baƒüƒ±mlƒ± deƒüi≈ükeni** (Y) ba≈üka **bir veya birka√ß baƒüƒ±msƒ±z deƒüi≈ükenle** (X) a√ßƒ±klamaya veya tahmin etmeye √ßalƒ±≈üan matematiksel modeldir.

### Ana unsurlar:
| Terim | A√ßƒ±klama |
|--------|----------|
| **Baƒüƒ±mlƒ± Deƒüi≈üken (Y)** | Tahmin edilmeye √ßalƒ±≈üƒ±lan deƒüi≈üken ‚Üí √ñrnek: Sƒ±nav notu, satƒ±≈ü fiyatƒ±, gelir |
| **Baƒüƒ±msƒ±z Deƒüi≈üken(ler) (X)** | Kullanƒ±lan a√ßƒ±klayƒ±cƒ± deƒüi≈üken(ler) ‚Üí √ñrnek: √áalƒ±≈üma s√ºresi, ev boyutu, reklam b√ºt√ßesi |
| **Katsayƒ±lar (Œ≤‚ÇÄ, Œ≤‚ÇÅ...):** | Modeldeki sabit terim ve eƒüimleri ifade eder |
| **Hata Terimi (Œµ):** | Modelin a√ßƒ±klayamadƒ±ƒüƒ± kƒ±sƒ±m |

üìå Basit doƒürusal regresyon denklemi:
$$
Y = \beta_0 + \beta_1 X + \varepsilon
$$


## üí° Ger√ßek Hayat √ñrneƒüi: Ev Fiyatlarƒ±

Bir emlak ≈üirketi, ev fiyatlarƒ±nƒ± ≈üu deƒüi≈ükenlerle tahmin etmek istiyor:

- Metrekare (X‚ÇÅ)
- Oda sayƒ±sƒ± (X‚ÇÇ)
- Konum skoru (X‚ÇÉ)

üëâ Bu durumda kullanacaƒüƒ±mƒ±z model:
$$
\text{Ev Fiyatƒ±} = \beta_0 + \beta_1 \cdot \text{Metrekare} + \beta_2 \cdot \text{Oda Sayƒ±sƒ±} + \beta_3 \cdot \text{Konum Skoru} + \varepsilon
$$

üìå Eƒüer metrekare artarsa, fiyat da artƒ±yor mu? Ya konum ne kadar etkili?

## üéØ Neden Regresyon Kullanƒ±lƒ±r?

| Ama√ß | A√ßƒ±klama |
|------|----------|
| **Tahmin Yapmak:** | √ñƒürencinin ka√ß alacaƒüƒ±nƒ±, maƒüazada ka√ß √ºr√ºn satƒ±lacaƒüƒ±nƒ±, hisse senedinin deƒüerini tahmin etmek i√ßin |
| **ƒ∞li≈ükiyi Anlamak:** | Hangi fakt√∂r√ºn daha √ßok etkili olduƒüunu g√∂rmek i√ßin |
| **Karar Almak:** | Veriye dayalƒ± stratejiler olu≈üturmak i√ßin |

üìå Yani regresyon, **veriyle geleceƒüi tahmin etmenin** en g√º√ßl√º yollarƒ±ndan biridir.

## üßÆ Regresyon T√ºrleri

| T√ºr | A√ßƒ±klama | Ne zaman kullanƒ±lƒ±r? |
|------|----------|-----------------------|
| **Basit Doƒürusal Regresyon** | 1 baƒüƒ±msƒ±z deƒüi≈üken | Tek bir √∂zellikle ili≈üki kurulacaksa |
| **√áoklu Doƒürusal Regresyon** | Birden fazla baƒüƒ±msƒ±z deƒüi≈üken | Birden fazla etken varsa |
| **Lojistik Regresyon** | Baƒüƒ±mlƒ± deƒüi≈üken evet/hayƒ±r ≈üeklinde olabilir | Sƒ±nƒ±flandƒ±rma yapmak i√ßin |
| **Polinom Regresyon** | ƒ∞li≈üki doƒürusal deƒüil, eƒüri gibi | Grafik doƒürusal deƒüilse |
| **Doƒürusal Olmayan Regresyon** | Karma≈üƒ±k ili≈ükiler varsa | Doƒürusal modeller yetmediƒüinde |


## üìà Basit Doƒürusal Regresyon Nasƒ±l √áalƒ±≈üƒ±r?

Varsayalƒ±m:
- Ders √ßalƒ±≈üma s√ºresi (X): [2, 3, 4, 5, 6] saat/g√ºn
- Sƒ±nav notu (Y): [50, 60, 70, 80, 90]

üëâ Hedef: **‚ÄúKa√ß saat √ßalƒ±≈üƒ±rsa ka√ß puan alƒ±r?‚Äù** sorusuna yanƒ±t vermek.

Model √ßƒ±ktƒ±sƒ± ≈ü√∂yle olabilir:
$$
\text{Not} = 30 + 10 \cdot (\text{√áalƒ±≈üma S√ºresi})
$$

üìå Bu form√ºle g√∂re:
- 2 saat √ßalƒ±≈üƒ±rsan: 30 + 10√ó2 = **50**
- 5 saat √ßalƒ±≈üƒ±rsan: 30 + 10√ó5 = **80**

Bu temel mantƒ±kla **hi√ß g√∂rmediƒüin bir ki≈üi i√ßin bile tahmin yapabilirsin.**

```python
# Python Uygulamasƒ±: Basit Doƒürusal Regresyon

import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Veri seti
x = np.array([2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([50, 60, 70, 80, 90])

# Regresyon modelini olu≈üturalƒ±m
model = LinearRegression()
model.fit(x, y)

# Eƒüim ve kesim noktasƒ±
print(f"Kesme (Œ≤‚ÇÄ): {model.intercept_:.2f}")
print(f"Eƒüim (Œ≤‚ÇÅ): {model.coef_[0]:.2f}")

# Tahmin yapalƒ±m
tahmin = model.predict(x)

# Grafiƒüe √ßizelim
plt.scatter(x, y, color='blue', label='Ger√ßek')
plt.plot(x, tahmin, color='red', label='Regresyon')
plt.title("Regresyon ile Tahmin")
plt.xlabel("√áalƒ±≈üma Saati")
plt.ylabel("Sƒ±nav Notu")
plt.legend()
plt.grid(True)
plt.show()
```
