## üìà Alan Grafiƒüi: Deƒüi≈üimi ve Katkƒ±yƒ± Aynƒ± Anda G√∂rmek

### Nedir?
Alan grafiƒüi, **√ßizgi grafik** ile **alan dolgusu** arasƒ±nda bir g√∂sterimdir. √ñzellikle zaman serisi verilerinde kullanƒ±lƒ±r.

### Ne ƒ∞≈üe Yarar?

- Farklƒ± kategorilerin toplam i√ßindeki katkƒ±sƒ±nƒ± vurgular

### T√ºrleri:
| T√ºr | A√ßƒ±klama |
|-----|-----------|
| **D√ºz Alan Grafiƒüi** | Her kategori baƒüƒ±msƒ±z g√∂sterilir |
| **Yƒ±ƒüƒ±lmƒ±≈ü Alan Grafiƒüi** | Toplam katkƒ±lar √ºst √ºste sƒ±ralanƒ±r |
| **%100 Yƒ±ƒüƒ±lmƒ±≈ü Alan Grafiƒüi** | Her zaman noktasƒ± %100‚Äôe g√∂re orantƒ±lanƒ±r ‚Üí oranlar √∂n planda |

### üí° Ger√ßek Hayat √ñrneƒüi:

Bir ≈üirketin aylƒ±k satƒ±≈ülarƒ±:

```
A √úr√ºn√º: [100, 120, 130, 150]  
B √úr√ºn√º: [80, 90, 110, 140]

```

üëâ Bu veriyi yƒ±ƒüƒ±lmƒ±≈ü alan grafiƒüine d√∂kersek:
- Hangi √ºr√ºn√ºn katkƒ± arttƒ±?

Bu sorulara tek bakƒ±≈üta yanƒ±t bulabilirsin.

## üìä √áubuk Grafik: Kategoriler Arasƒ± Farkƒ± G√∂steren Silah

√áubuk grafik, farklƒ± **kategorileri kar≈üƒ±la≈ütƒ±rmak** i√ßin kullanƒ±lan g√º√ßl√º bir g√∂rsel ara√ßtƒ±r.

### T√ºrleri:

### üí° Ger√ßek Hayat √ñrneƒüi:

üëâ Bu verileri gruplanmƒ±≈ü √ßubuk grafiƒüe d√∂kersen:

## ü•ß Pasta Grafiƒüi: Bir B√ºt√ºn√ºn Par√ßalarƒ±nƒ± G√∂stermek

Pasta grafiƒüi, bir b√ºt√ºn√º olu≈üturan par√ßalarƒ±n **oranlarƒ±nƒ±** g√∂sterebilen dairesel bir grafiktir.

## üìâ √áizgi Grafik: Trendleri Yakalamak ƒ∞√ßin En G√º√ßl√º Silah

### Nedir?
Zaman i√ßindeki deƒüi≈üimi g√∂stermek i√ßin kullanƒ±lan √ßizgiyle olu≈üturulan grafiktir.

### Ne ƒ∞≈üe Yarar?
- Hisse senedi fiyatlarƒ±nda dalgalanmalarƒ± g√∂rmek
- Haftalƒ±k satƒ±≈ülarda y√ºkselen/d√º≈üen eƒüilimleri yakalamak
- Sƒ±caklƒ±k deƒüi≈üimini zamana g√∂re incelemek

### T√ºrleri:
| T√ºr | A√ßƒ±klama |
|------|----------|
| **Basit √áizgi Grafik** | Tek bir veri serisini g√∂sterir |
| **√áoklu √áizgi Grafik** | Birden fazla seriyi aynƒ± grafikte kar≈üƒ±la≈ütƒ±rƒ±r |
| **Yƒ±ƒüƒ±lmƒ±≈ü √áizgi Grafik** | Seriler birbirinin √ºzerine eklenerek toplamƒ± g√∂sterir |

### üí° Ger√ßek Hayat √ñrneƒüi:

Aylƒ±k sƒ±caklƒ±k verisi:

```
Ocak: 5¬∞C  
≈ûubat: 7¬∞C  
Mart: 10¬∞C  
Nisan: 15¬∞C
```

üëâ Bu verileri √ßizgi grafiƒüe d√∂kt√ºƒü√ºnde:
‚ÄúSƒ±caklƒ±k artƒ±yor mu, yoksa rastgele mi deƒüi≈üiyor?‚Äù  
sorusuna kolayca yanƒ±t bulursun.


````python
# Python Uygulamasƒ±: Grafikler Nasƒ±l √áizilir?

import matplotlib.pyplot as plt
import numpy as np

# Basit √ßizgi grafik
x = ["Ocak", "≈ûubat", "Mart", "Nisan"]
y = [5, 7, 10, 15]

plt.plot(x, y, marker='o')
plt.title("Sƒ±caklƒ±k Artƒ±≈üƒ±")
plt.xlabel("Ay")
plt.ylabel("Sƒ±caklƒ±k (¬∞C)")
plt.grid(True)
plt.show()
````

````python
# Yƒ±ƒüƒ±lmƒ±≈ü Alan Grafiƒüi

import pandas as pd

data = {
    'A': [100, 120, 130, 150],
    'B': [80, 90, 110, 140],
    'C': [60, 70, 80, 100]
}
df = pd.DataFrame(data, index=["Ocak", "≈ûubat", "Mart", "Nisan"])

df.plot(kind='area', stacked=True)
plt.title("Yƒ±ƒümalƒ± Alan Grafiƒüi")
plt.xlabel("Ay")
plt.ylabel("Satƒ±≈ülar")
plt.show()
````

````python
# Gruplandƒ±rƒ±lmƒ±≈ü √áubuk Grafik

import numpy as np

N = 4
ind = np.arange(N)  # x ekseni pozisyonlarƒ±
width = 0.2        # √ßubuk geni≈üliƒüi

fig, ax = plt.subplots()

rects1 = ax.bar(ind, df['A'], width, label='A √úr√ºn√º')
rects2 = ax.bar(ind + width, df['B'], width, label='B √úr√ºn√º')
rects3 = ax.bar(ind + 2*width, df['C'], width, label='C √úr√ºn√º')

ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Ocak', '≈ûubat', 'Mart', 'Nisan'))
ax.legend()
plt.title('Aylara G√∂re √úr√ºn Satƒ±≈ülarƒ±')
plt.show()
````

````python
# Pasta Grafiƒüi

gelir_kaynaklari = [50, 30, 15, 5]
etiketler = ['√úr√ºn Satƒ±≈üƒ±', 'Reklam', 'Sponsorluk', 'Diƒüer']

plt.pie(gelir_kaynaklari, labels=etiketler, autopct='%1.1f%%')
plt.title("Gelir Kaynaklarƒ±")
plt.axis('equal')  # Pasta grafiƒüini d√ºzg√ºn g√∂ster
plt.show()
````
